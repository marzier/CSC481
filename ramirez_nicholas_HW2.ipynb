```python
#data frame

import pandas as pd

data = {
    "name": ['alice', 'bob', 'charlie', 'david'],
    'age': [25,30,17,28],
    'city': ['new york', 'san francisco', 'los angeles', 'alexandria'],
    'salary': [100, 150, 200, 300]
}

df = pd.DataFrame(data)

df[['name', 'salary']]

df.iloc[0]
```

```python
# calculate some statistics on each columbn
df['age'].mean()
```

Series

```python
print("the minimum salary is", df['salary'].min())

print("the mean salary is", df['salary'].mean())

print("the max slaary is", df['salary'].max())
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,15,7,12,8]

# create a line plot
plt.plot(x,y, marker='o', markersize=8, color='red', linestyle='--', linewidth=2)

#add labels and titles
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('x versus y')
plt.grid(True)

plt.show()

```

```python
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales':  [500, 800, 1700, 1500, 2000]
}

df = pd.DataFrame(data)

#create a line plot
plt.plot(df['Year'], df['Sales'],marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Years')
plt.grid(True)

#set x-axis ticks at integer values
plt.xticks(df['Year'])
plt.show()
```

```python
data = {
    'Category': ['A','B','C','D'],
    'Sales': [500,800,1200,1300]
}

df = pd.DataFrame(data)

# create a bar plot
plt.bar(df['Category'], df['Sales'], color='grey')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()

```

```python
# create a DataFrame

data = {
    'Age' : [25,30,35,40,45],
    'Income' : [50000,60000,4500,70000,55000]
}

df = pd.DataFrame(data)

plt.scatter(df['Age'], df['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Income vs Age')
plt.show()
```

```python
# creating data frame
data = {
    'Scores': [85,78,92,65,80,75,88,90,82,70]
}

df = pd.DataFrame(data)

# create a histogram of data
plt.hist(df['Scores'], width=1.5)
plt.xlabel('scores')
plt.ylabel('frequency')
plt.title('dist of scores')
plt.show()
```
